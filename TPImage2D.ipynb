```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, exposure, transform
import scipy.ndimage as ndi
```

```python
image = io.imread('https://www.bigspeak.com/wp-content/uploads/2014/12/Kevin-Mitnick-crop-scaled-e1591831035717.jpg')
image_gray = color.rgb2gray(image)
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Image originale')
plt.imshow(image, cmap='gray')
```

```python
plt.subplot(1, 2, 2)
plt.title('Image en niveaux de gris')
plt.imshow(image_gray, cmap='gray')
plt.show()
```

```python
taille_standard = (100, 100)
image_resized = transform.resize(image_gray, taille_standard)
```

```python
plt.figure()
plt.title('Image redimensionnée')
plt.imshow(image_resized, cmap='gray')
plt.show()
```

```python
image_eq = exposure.equalize_hist(image_resized)
```

```python
plt.figure()
plt.title('Image avec histogramme égalisé')
plt.imshow(image_eq, cmap='gray')
plt.show()
```

```python
image_smooth = ndi.gaussian_filter(image_eq, sigma=1)
```

```python
plt.figure()
plt.title('Image après lissage')
plt.imshow(image_smooth, cmap='gray')
plt.show()
```
